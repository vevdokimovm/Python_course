```python
import requests

api_url = "http://numbersapi.com"

params = {
    'default': 'Boring',
    'json': 'true'
}

with open("dataset.txt", "r") as infile:
    numbers = infile.read().splitlines()
    
# print(numbers)

for number in numbers:
#     url = f'{api_url}/{number}/math'
#     res = requests.get(url, params=params)
#     data = res.json()
# #     print(data)
#     if data['text'] == 'Boring':
#         print('Boring')
#     else:
#         print('Interesting')
    
    number = number.strip()
    url = api_url + '/' + number
    res = requests.get(url, params=params)
    print(url)
    print(res.status_code)
    data = res.json()
    print(data)
    print("text = ", data['text'])
    if data['text'] == 'Boring':
        print('Boring')
    else:
        print('Interesting')
        
    
```
